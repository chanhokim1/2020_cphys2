# logistic regression

iris data의 target은 3개이기 때문에 logistic regression으로 3개를 동시에 분류하는 것은 불가능하다.\
logistic regression이 setosa, versicolor, virginica를 각각 얼마나 잘 분류할 수 있는지는 알아볼 수 있다.


```python
from sklearn.linear_model import LogisticRegression
regress = LogisticRegression(solver='lbfgs')
```

```python
SL = [[i] for i in iris.data[:,0]] # 꽃받침 길이
SW = [[i] for i in iris.data[:,1]] # 꽃받침 넓이
PL = [[i] for i in iris.data[:,2]] # 꽃잎 길이
PW = [[i] for i in iris.data[:,3]] # 꽃잎 넓이
```

### 1. setosa 분류

```python
Y_setosa = [1 if i==0 else 0 for i in iris.target]
```

#### 1 - (1) sepal length

```python
# 꽃받침 길이
regress.fit(SL, Y_setosa)
print('Score of sepal length :', regress.score(SL,Y_setosa))
XSL=[[i] for i in np.linspace(4,8,100)]
plt.plot(SL, Y_setosa, 'o')
plt.plot(XSL, regress.predict(XSL))
plt.title('sepal length - setosa')
```

#### 1 - (2) sepal width

```python
# 꽃받침 넒이
regress = LogisticRegression(solver='lbfgs')
regress.fit(SW, Y_setosa)
print('Score of sepal width :', regress.score(SW,Y_setosa))
XSW=[[i] for i in np.linspace(1.5,5,100)]
plt.plot(SW, Y_setosa, 'o')
plt.plot(XSW, regress.predict(XSW))
plt.title('sepal width - setosa')
```

#### 1 - (3) petal length

```python
# 꽃잎 길이
regress = LogisticRegression(solver='lbfgs')
regress.fit(PL, Y_setosa)
print('Score of petal length :', regress.score(PL,Y_setosa))
XPL=[[i] for i in np.linspace(1,8,100)]
plt.plot(PL, Y_setosa, 'o')
plt.plot(XPL, regress.predict(XPL))
plt.title('petal length - setosa')
```

#### 1 - (4) petal width

```python
# 꽃잎 넒이
regress = LogisticRegression(solver='lbfgs')
regress.fit(PW, Y_setosa)
print('Score of petal width :', regress.score(PW,Y_setosa))
XPW=[[i] for i in np.linspace(0,3,100)]
plt.plot(PW, Y_setosa, 'o')
plt.title('petal width - setosa')
plt.plot(XPW, regress.predict(XPW))
```

꽃받침으로 구한 score는 0.89, 0.83이지만 꽃잎으로 구한 score는 모두 1.0이다.\
전체적으로 꽃받침 보다는 꽃잎으로 분류를 하면 더 정확히 분류를 할 수 있다.

## Supprt Vector Machine

support vector machine을 이용하면 2개 보다 더 많은 lable을 한번에 분류할 수 있다.

```python
from sklearn.svm import SVC
y = iris.target
```

```python
X=iris["data"] 
y=iris["target"] # setosa=0, versicolor=1, virginica=2
```

### 1. linear kernal 분류

```python
svm=SVC(kernel='linear',gamma='scale')
svm.fit(X,y)
svm.score(X,y)
```

```python
np.shape(svm.decision_function(X))
```

```python
for i in range(len(iris.feature_names)):
    for j in range(len(iris.feature_names)):
        if j != i:
            plt.scatter(X[:,i],X[:,j],c=svm.predict(X))
            plt.xlabel(iris.feature_names[i])
            plt.ylabel(iris.feature_names[j])
            plt.show()
        
```

### 2. rbf kernal 분류

```python
svm=SVC(kernel='rbf',gamma='scale')
svm.fit(X,y)
svm.score(X,y)
```

```python
for i in range(len(iris.feature_names)):
    for j in range(len(iris.feature_names)):
        if j != i:
            plt.scatter(X[:,i],X[:,j],c=svm.predict(X))
            plt.xlabel(iris.feature_names[i])
            plt.ylabel(iris.feature_names[j])
            plt.show()
```

### 3. sigmoid kernal 분류

```python
svm=SVC(kernel='sigmoid',gamma='scale')
svm.fit(X,y)
svm.score(X,y)
```

```python
for i in range(len(iris.feature_names)):
    for j in range(len(iris.feature_names)):
        if j != i:
            plt.scatter(X[:,i],X[:,j],c=svm.predict(X))
            plt.xlabel(iris.feature_names[i])
            plt.ylabel(iris.feature_names[j])
            plt.show()
```

iris data의 features는 4개, lable은 3개로 하나의 그래프 안에 모두 표기할 수 없다.\
따라서 2가지 특성을 하나씩 골라서 x축과 y축에 넣어 2차원 평면상으로 투영한 그림이다.

linear kernal과 rbf kernal은 높은 정확도로 iris를 분류하지만 sigmoid kernal은 제대로 분류할 수 없었다.
