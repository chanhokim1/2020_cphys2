## Boston House price

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
x = boston.data    # features
y = boston.target  # labels

name = boston.feature_names # name of features
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regress = LinearRegression()
for i in range(len(name)):
    feature = np.atleast_2d(x[:,i])
    X = np.transpose(feature)
    regress.fit(X,y)
    fea_name = str(name[i])
    print(name[i])
    print('coef', regress.coef_)
    print('intercept', regress.intercept_)
    print('score', regress.score(X,y))
    plt.plot(X, regress.predict(X))
    plt.plot(X,y,'o')
    plt.title(fea_name)
    plt.show()
```

각각의 features로 coef와 intercept로 regression을 한 결과 RM, LSTAT 적합도(score)가 가장 높다.

Lasso regression를 적용하면 결과는 다음과 같다.

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=10)
for i in range(len(name)):
    feature = np.atleast_2d(x[:,i])
    Xl = np.transpose(feature)
    lasso.fit(Xl,y)
    fea_name = str(name[i])
    print(name[i])
    print('coef',lasso.coef_)
    print('intercept',lasso.intercept_)
    print('score', lasso.score(Xl,y))
    print('')
    plt.plot(Xl,lasso.predict(Xl))
    plt.plot(Xl,y,'o')
    plt.title(fea_name)
    plt.show()
```

Ridge regression을 적용하면 다음과 같다.

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=10)
for i in range(len(name)):
    feature = np.atleast_2d(x[:,i])
    Xr = np.transpose(feature)
    ridge.fit(Xr,y)
    fea_name = str(name[i])
    print(name[i])
    print(ridge.coef_)
    print(ridge.intercept_)
    print(ridge.score(Xr,y))
    print('')
    plt.plot(Xr,ridge.predict(Xr))
    plt.plot(Xr,y,'o')
    plt.title(fea_name)
    plt.show()
```

coef와 intercept는 변하지만 score는 3가지 regression에서 거의 일정하다.
