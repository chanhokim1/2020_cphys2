# image Segmentation

수업시간에 실습한 예제를 따라서 다음 사진에서 색상별로 분리하여 어떻게 보이는지 확인하기.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_image
from sklearn.cluster import KMeans

china = load_sample_image("china.jpg")
image = china /255
plt.imshow(image)
```

```python
china # 각 픽셀의 RGB 값. 1부터 255까지 있음.
```

```python
# 각각의 픽셀에서 RGB 값을 255로 나눈 값
image 
```

```python
X = image.reshape(-1,3) # image 리스트는 3차원 배열인데, 이 배열을 2차원 배열로 변경. 
print('image',np.shape(image))
print('X', np.shape(X))
```

```python
kmeans = KMeans(n_clusters=8).fit(X) # 비슷한 색상끼리 clustering을 함. 총 몇개의 cluster로 나눌지 설정
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img) # 총 8개의 색상으로 나눔. 8개 색상이 사용됨.
```

```python
kmeans = KMeans(n_clusters=4).fit(X) # 4개의 군집으로 나눈 경우,
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img) # 총 4개의 색상으로 나눔. 4개 색상이 사용됨.
```
